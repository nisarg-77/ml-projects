```python
data['Hour'] = data['Time'].apply(lambda x : np.ceil(float(x)/3600) % 24)
data['Minute'] = data['Time'].apply(lambda x: np.ceil(float(x)/60)%60)
```

### Their is clear distinction in average value of both classes

1. The **average hour** for fraudulent transactions is around **`midday (12 PM)`**.
2. Fraudulent transactions are spread throughout the day, with a standard deviation of **`6.66 hours`**.
3. The **25%** of fraudulent transactions happen before **`5 AM`**, while **75%** occur before **`5 PM`**.

4. The **average hour** for non fraudulent transactions is around **`afternoon (2 PM)`**.
5. Non fraudulent transactions are spread throughout the day, with a standard deviation of **`5.83 hours`**.
6. The **25%** of non fraudulent transactions happen before **`10 AM`**, while **75%** occur before **`7 PM`**.

```python
def neural_net_predictions(model, x):
  return (model.predict(x).flatten() > 0.5).astype(int)
neural_net_predictions(shallow_nn, X_test)
```

```python
for i in range(5,6):
    for j in range(i, 5*i):
        data_exp_fraud = data_fraud.sample(count_fraud*i, replace = True)
        data_exp_not_fraud = data_not_fraud.sample(count_fraud*j)
        data_exp = pd.concat([data_exp_fraud,data_exp_not_fraud], axis = 'rows')
        
        X_exp = data_exp.drop("Class", axis = 'columns')
        y_exp = data_exp.Class
        X_exp_train, X_exp_test, y_exp_train, y_exp_test = train_test_split(X_exp, y_exp, test_size = 0.2, random_state = 17, stratify = y_exp)
        
        rf_5 = RandomForestClassifier(max_depth = 6, n_jobs = -1)
        rf_5.fit(X_exp_train, y_exp_train)
        
        print(i,j)
        print(classification_report(y_test, rf_5.predict(X_test), target_names=['Not Fraud', 'Fraud']))
        roc_auc = roc_auc_score(y_test, rf_5.predict(X_test))
        print(f"\nROC-AUC Score: {roc_auc}")
```

### Hybrid Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
# Calculate class distribution before Sampling
print("Class distribution before Sampling:")
print(Counter(y_train))

# Initialize RandomUnderSampler and apply it to the training data
rus = RandomUnderSampler(sampling_strategy=0.003)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

# Calculate class distribution after random under-sampling
print("Class distribution after RandomUnderSampler:")
print(Counter(y_resampled))

# Initialize RandomOverSampler and apply it to the resampled data
ros = RandomOverSampler(sampling_strategy=0.08)
X_resampled, y_resampled = ros.fit_resample(X_resampled, y_resampled)

print("Class distribution after RandomOverSampler:")
print(Counter(y_resampled))
```
